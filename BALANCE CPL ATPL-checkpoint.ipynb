```python
import pandas as pd
import os
```

```python
student_name = input('Inserte el nombre del estudiante (APELLIDO Nombre): ')
base_path = 'C:\\Users\\confi\\Richporter Lighting\\Operations Select - Documents\\01. Airplane\\03. Licenses Emissions (A)\\02. Airplane Students Records\\'
student_path = os.path.join(base_path, student_name,'CPL BALANCE',f'PTR-{student_name}.xlsx')
df = pd.read_excel(student_path, header=None)
df
```

```python
new_row = ['Date','Type','Registration','Pic/Copic','Description','From','To','SE_day_dual','SE_day_pic','SE_night_dual','SE_night_pic','ME_day_dual','ME_day_pic','ME_day_copic','ME_night_dual','ME_night_pic','ME_night_copic','IMC','Hood','FTD','ifr appr','x-country_day_dual','x_country_day_pic','x-country_night_dual','x-country_night_pic','T/O_day','T/O_night','VFR_FTD','Theory','PGI','Studies','Exams']
new_row_df = pd.DataFrame([new_row])
df = df.reset_index(drop=True)
new_row_df = new_row_df.reset_index(drop=True)
df = pd.concat([new_row_df,df], ignore_index = True)
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)
df = df.iloc[9:].reset_index(drop=True)
df = df.iloc[:, :-4]
df
```

```python
#df filtrados
```

```python
df_ppl = df[df['Description'].str.contains('PPL', na=False)]
df_ft_ppl = df_ppl[df_ppl['Description'].str.contains('FT', na=False)]
df_vfr = df[df['Description'].str.contains('S3',na=False) & ~df['Type'].str.contains('PA44',na=False)]
df_cpl = df[df['Description'].str.contains('S4',na=False)]
df_ft_cpl = df_cpl[df_cpl['Description'].str.contains('FT', na=False)]
df_multi = df[df['Description'].str.contains('IRME',na=False) | df['Description'].str.contains('Multi',na=False)]
df_multi_ft = df[df['Description'].str.contains('IRME',na=False) &  df['Description'].str.contains('FT',na=False)| df['Description'].str.contains('Multi',na=False) &  df['Description'].str.contains('FT',na=False)]
df_c172 = df[df['Type'].str.contains('C172',na=False)]
df_int = df[df['Description'].str.contains('S5',na=False)]
```

```python
#calcula fase PPL
```

```python
dual_hours_ppl = round(df_ppl['SE_day_dual'].sum() + df_ppl['SE_night_dual'].sum() - df_ft_ppl['SE_day_dual'].sum(),2)
solo_hours_ppl = round(df_ppl['SE_day_pic'].sum() + df_ppl['SE_night_pic'].sum(),2)
ft_hours_ppl = round(df_ft_ppl['SE_day_dual'].sum(),2)
sim_hours_ppl = round(df_ppl['VFR_FTD'].sum() + df_ppl['FTD'].sum(),2)
print('In the main PTR, PPL phase: ')
print(f'{dual_hours_ppl} hours of duals')
print(f'{solo_hours_ppl} hours of solo')
print(f'{ft_hours_ppl} hours of Flight Test')
print(f'{sim_hours_ppl} hours of simulator')
```

```python
#calcula fase VFR
```

```python
dual_hours_vfr = round(df_vfr['SE_day_dual'].sum() + df_vfr['SE_night_dual'].sum(),2)
solo_hours_vfr = round(df_vfr['SE_day_pic'].sum() + df_vfr['SE_night_pic'].sum(),2)
sim_hours_vfr = round(df_vfr['VFR_FTD'].sum() + df_vfr['FTD'].sum(),2)
dual_hours_ifr = round(df_multi['ME_day_dual'].sum() + df_multi['ME_night_dual'].sum()-df_multi_ft['ME_day_dual'].sum(),2)
ft_ifr_hours = round(df_multi_ft['ME_day_dual'].sum(),2)
print('In the main PTR, VFR phase: ')
print(f'{dual_hours_vfr} hours of duals')
print(f'{solo_hours_vfr} hours of solo')
print(f'{sim_hours_vfr} hours of simulator')
print(f'{dual_hours_ifr} hours of duals in PA44')
print(f'{ft_ifr_hours} hours of Flight Test in PA44')
```

```python
#Calcula Fase CPL
```

```python
dual_hours_cpl = round(df_cpl['SE_day_dual'].sum() + df_cpl['SE_night_dual'].sum() - df_ft_cpl['SE_day_dual'].sum(),2)
solo_hours_cpl = round(df_cpl['SE_day_pic'].sum() + df_cpl['SE_night_pic'].sum(),2)
ft_hours_cpl = round(df_ft_cpl['SE_day_dual'].sum(),2)
print('In the main PTR, CPL phase: ')
print(f'{dual_hours_cpl} hours of duals')
print(f'{solo_hours_cpl} hours of solos')
print(f'{ft_hours_cpl} hours of Flight Test')
```

```python
solo_hours_int = round(df_int['SE_day_pic'].sum()+ df_int['SE_night_pic'].sum(),2)
sim_int = round(df_int['VFR_FTD'].sum()+df_int['FTD'].sum(),2)
print('In the main PTR, integrated phase: ')
print(f'{solo_hours_int} hours of solo')
print(f'{sim_int} hours of sim')
```
