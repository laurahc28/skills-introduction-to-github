```python
import pandas as pd
import os
```

```python
student_name = input('Inserte el nombre del estudiante (APELLIDO Nombre): ')
base_path = 'C:\\Users\\confi\\Richporter Lighting\\Operations Select - Documents\\01. Airplane\\03. Licenses Emissions (A)\\02. Airplane Students Records\\'
student_path = os.path.join(base_path, student_name,'CPL BALANCE',f'PTR-{student_name}.xlsx')
df = pd.read_excel(student_path, header=None)
df
```

```python
new_row = ['Date','Type','Registration','Pic/Copic','Description','From','To','SE_day_dual','SE_day_pic','SE_night_dual','SE_night_pic','ME_day_dual','ME_day_pic','ME_day_copic','ME_night_dual','ME_night_pic','ME_night_copic','IMC','Hood','FTD','ifr appr','x-country_day_dual','x_country_day_pic','x-country_night_dual','x-country_night_pic','T/O_day','T/O_night','VFR_FTD','Theory','PGI','Studies','Exams']
new_row_df = pd.DataFrame([new_row])
df = df.reset_index(drop=True)
new_row_df = new_row_df.reset_index(drop=True)
df = pd.concat([new_row_df,df], ignore_index = True)
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)
df = df.iloc[9:].reset_index(drop=True)
df = df.iloc[:, :-4]
df
```

```python
#df filtrados
```

```python
df_ppl = df[df['Description'].str.contains('PPL', na=False)]
df_ft_ppl = df_ppl[df_ppl['Description'].str.contains('FT', na=False)]
df_ft = df[df['Description'].str.contains('FT', na=False)]
df_cpl = df[df['Description'].str.contains('NR',na=False) | df['Description'].str.contains('CPL',na=False) | df['Description'].str.contains('TB',na=False)]
df_multi = df[df['Description'].str.contains('IFR',na=False) | df['Description'].str.contains('Multi',na=False)]
df_c172 = df[df['Type'].str.contains('C172',na=False)]
```

```python
#calcula fase PPL
```

```python
dual_hours_ppl = round(df_ppl['SE_day_dual'].sum() + df_ppl['SE_night_dual'].sum() - df_ft_ppl['SE_day_dual'].sum(),2)
solo_hours_ppl = round(df_ppl['SE_day_pic'].sum() + df_ppl['SE_night_pic'].sum(),2)
ft_hours_ppl = round(df_ft_ppl['SE_day_dual'].sum(),2)
sim_hours_ppl = round(df_ppl['VFR_FTD'].sum() + df_ppl['FTD'].sum(),2)
print('En el PTR principal Fase PPL: ')
print(f'Tiene {dual_hours_ppl} horas de duales')
print(f'Tiene {solo_hours_ppl} horas de solo')
print(f'Tiene {ft_hours_ppl} horas de Flight Test')
print(f'Tiene {sim_hours_ppl} horas de simulador')
```

```python
#calcula fase CPL
```

```python
dual_hours_cpl = round(df_cpl['SE_day_dual'].sum() + df_cpl['SE_night_dual'].sum() ,2)
solo_hours_cpl = round(df_cpl['SE_day_pic'].sum() + df_cpl['SE_night_pic'].sum(),2)
ft_hours_cpl = round(df_ft['SE_day_dual'].sum() - ft_hours_ppl,2)
sim_hours_cpl = round(df['VFR_FTD'].sum() + df['FTD'].sum() - sim_hours_ppl - df_multi['VFR_FTD'].sum() - df_multi['FTD'].sum(),2)
print('En el PTR principal Fase CPL: ')
print(f'Tiene {dual_hours_cpl} horas de duales')
print(f'Tiene {solo_hours_cpl} horas de solo')
print(f'Tiene {ft_hours_cpl} horas de Flight Test')
print(f'Tiene {sim_hours_cpl} horas de simulador')
```

```python
#Calcula Multi
```

```python
dual_hours_ifr = round(df['ME_day_dual'].sum() + df['ME_night_dual'].sum(),2)
solo_hours_ifr = round(df['ME_day_pic'].sum() + df['ME_night_pic'].sum(),2)
sim_hours_ifr = round(df_multi['VFR_FTD'].sum() + df_multi['FTD'].sum(),2)
print('En el PTR principal Fase Multi/IFR: ')
print(f'Tiene {dual_hours_ifr} horas de duales')
print(f'Tiene {solo_hours_ifr} horas de solo')
print(f'Tiene {sim_hours_ifr} horas de simulador')
```
