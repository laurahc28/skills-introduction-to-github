```python
import pandas as pd
import os
```

```python
student_name = input('Inserte el nombre del estudiante: ')
base_path = 'C:\\Users\\confi\\Richporter Lighting\\Operations Select - Documents\\01. Airplane\\03. Licenses Emissions (A)\\02. Airplane Students Records\\'
student_path = os.path.join(base_path, student_name,'PPL',f'PTR-{student_name}.xlsx')
df = pd.read_excel(student_path, header=None)
df
```

```python
new_row = ['Date','Type','Registration','Pic/Copic','Description','From','To','SE_day_dual','SE_day_pic','SE_night_dual','SE_night_pic','ME_day_dual','ME_day_pic','ME_day_copic','ME_night_dual','ME_night_pic','ME_night_copic','IMC','Hood','FTD','ifr appr','x-country_day_dual','x_country_day_pic','x-country_night_dual','x-country_night_pic','T/O_day','T/O_night','VFR_FTD','Theory','PGI','Studies','Exams']
new_row_df = pd.DataFrame([new_row])
df = df.reset_index(drop=True)
new_row_df = new_row_df.reset_index(drop=True)
df = pd.concat([new_row_df,df], ignore_index = True)
df.columns = df.iloc[0]
df = df.drop(0).reset_index(drop=True)
df = df.iloc[9:].reset_index(drop=True)
df = df.iloc[:, :-4]
df
```

```python
df_ppl = df[df['Description'].str.contains('PPL', na=False)]
```

```python
student_path2 = os.path.join(base_path, student_name,'PPL',f'PTR ADD-{student_name}.xlsx')
df2 = pd.read_excel(student_path2, header = None)
new_row_df2 = pd.DataFrame([new_row])
df2 = df2.reset_index(drop=True)
new_row_df2 = new_row_df2.reset_index(drop=True)
df2 = pd.concat([new_row_df2,df2], ignore_index = True)
df2.columns = df2.iloc[0]
df2 = df2.drop(0).reset_index(drop=True)
df2 = df2.iloc[9:].reset_index(drop=True)
df2
```

```python
df_ppl_add = df2[df2['Description'].str.contains('PPL', na=False)]
```

```python
total_flight_time = round(df_ppl['SE_day_dual'].sum() + df_ppl['SE_night_dual'].sum() + df_ppl_add['SE_day_dual'].sum()+df_ppl_add['SE_night_dual'].sum()+df_ppl['SE_day_pic'].sum() + df_ppl['SE_night_pic'].sum() + df_ppl_add['SE_day_pic'].sum()+ df_ppl_add['SE_night_pic'].sum(),2)
dual_hours = round(df_ppl['SE_day_dual'].sum() + df_ppl['SE_night_dual'].sum() + df_ppl_add['SE_day_dual'].sum()+df_ppl_add['SE_night_dual'].sum(),2)
solo_hours = round(df_ppl['SE_day_pic'].sum() + df_ppl['SE_night_pic'].sum() + df_ppl_add['SE_day_pic'].sum()+ df_ppl_add['SE_night_pic'].sum(),2)
dual_cross_country = round(df_ppl['x-country_day_dual'].sum() + df_ppl['x-country_night_dual'].sum()+df_ppl_add['x-country_day_dual'].sum()+df_ppl_add['x-country_night_dual'].sum(),2)
solo_cross_country = round(df_ppl['x_country_day_pic'].sum() + df_ppl['x-country_night_pic'].sum()+df_ppl_add['x_country_day_pic'].sum()+df_ppl_add['x-country_night_pic'].sum() ,2)
ins_flight_time = round(df_ppl['IMC'].sum() + df_ppl['Hood'].sum()+df_ppl_add['IMC'].sum() + df_ppl_add['Hood'].sum(),2)
ground_flight_time = round(df_ppl['FTD'].sum()+df_ppl_add['FTD'].sum(),2)
print('En el PTR: ')
print(f'Tiene {total_flight_time} horas de total flight time')
print(f'Tiene {dual_hours} horas de duales')
print(f'Tiene {solo_hours} horas de solo')
print(f'Tiene {dual_cross_country} horas de duales en cross-country')
print(f'Tiene {solo_cross_country} horas de solo en cross-country')
print(f'Tiene {ins_flight_time} horas de instrument time')
print(f'Tiene {ground_flight_time} horas de ground time')
```
